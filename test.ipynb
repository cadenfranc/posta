```python
hashtag_performance = dict()
for post in basic_insights['data']:
    hashtags = [word for word in post["caption"].split() if word[0] == "#"]
    for hashtag in hashtags:
        try:
            hashtag_performance[hashtag]["likes"] += post["like_count"]
            hashtag_performance[hashtag]["usage"] += 1
        except KeyError:
            hashtag_performance[hashtag] = {
                "likes": post["like_count"],
                "usage": 1,
            }
            
pd.Series(hashtag_performance)
```
